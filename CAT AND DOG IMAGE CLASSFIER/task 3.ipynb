# Importing the packages and libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Reading and exploring the data

```python
df=pd.read_csv("C:\\Users\\daggu\\Downloads\\cats_vs_dogs.csv")
```

```python
df.rename(columns={'Unnamed: 0':'Index'},inplace=True)
df.head()
```

```python
df.head(10)
```

```python
df.set_index('Index',inplace=True)
df.head()
```

```python
sns.regplot(data=df,x='n_households',y='n_pet_households',)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='state',y='n_pet_households')
plt.xticks(rotation=75)
plt.ylabel('Number of Pet Households')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='state',y='n_dog_households')
plt.xticks(rotation=75)
plt.ylabel('Number of Dog Households')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='state',y='n_cat_households')
plt.xticks(rotation=75)
plt.ylabel('Number of Cat Households')
plt.show()
```

```python
Dog_pop=df['dog_population'].sum()
Cat_pop=df['cat_population'].sum()
Total_pop= Dog_pop+Cat_pop
print('Among '+str(Total_pop)+' pets '+ str(Dog_pop)+' are Dogs and '+ str(Cat_pop)+' are Cats')
```

```python
df1=df[['state','cat_population','dog_population']]
df1.head()
```

```python
df1.set_index('state',inplace=True)
```

```python
df1.transpose()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df1,fmt='d', annot=True,  cmap='YlGnBu', linewidths=0.5)
plt.title('Cat and Dog Populations by State')
plt.show()
```
